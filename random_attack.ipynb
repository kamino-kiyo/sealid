```python

### 英15文字をランダムに生成して現役IDに当たるまで試行

import importnb
with __import__('importnb').Notebook():
    import func as f

# 攻撃チャレンジ
def attack_seal_id():
    count = 0
    attacked_list = []

    # TODO:最終的には無限ループにしたいけど一旦回数制限付きで試す
    while count < 10000:

        # まだ攻撃を試してないランダムな文字列を生成
        while True:
            if (attacking_id := f.create_random_basetable_str(15)) in attacked_list:
                continue
            else:
                break

        if f.is_exist_seal_id(attacking_id):
            print(attacking_id)
            print(count)
            break
        else:
            attacked_list.append(attacking_id) 
            count += 1
        
    print("end")

attack_seal_id()
```
