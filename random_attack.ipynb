```python
import random

# シールIDに使っている文字列をランダムに並び替えるテーブル
BASE_TABLE = ["F","S","L","G","J","X","R","K","C","I","Q","Z","V","A","M","W","U","N","T","B","E","H","P","O","Y","D"]

# 26進数(英字)表記を10進数に変換
def id_to_decimal(src):
    # 下の桁から計算したいので文字列を逆順にする
    reverse_src = ''.join(list(reversed(src)))
    
    dst = 0
    count = 0
    for letter in reverse_src:
        dst += BASE_TABLE.index(letter) * len(BASE_TABLE) ** count
        count += 1
    return dst

# 存在するシールIDかどうかのチェック関数
def is_exist_seal_id(id):
    # 入力文字列を10進数に変換
    decimal_id = id_to_decimal(id)

    # チェックサムが正常かどうか
    id_data = str(decimal_id)[:-2]
    checksum = str(decimal_id)[-2:]

    sum = 0
    for letter in str(decimal_id)[:-2]:
        sum += int(letter)
    
    print("==========")
    print(decimal_id)
    print(id_data)
    print(checksum)
    print(sum)
    
    if sum != checksum:
        return False
    
    # 先頭10文字を日付に変換して、直近14日以内かどうか

    # 次の5文字が機器IDかどうか

    # 全部OKならシールID判定
    return True

# BASE_TABLEの内容をランダムに使って文字列を生成する
def create_random_basetable_str(length):
    return "".join([BASE_TABLE[random.randrange(len(BASE_TABLE))] for length in range(length)])

# 攻撃チャレンジ
def attack_seal_id():
    count = 0
    attacked_list = []

    # TODO:最終的には無限ループにしたいけど一旦回数制限付きで試す
    while count < 10000:

        # まだ攻撃を試してないランダムな文字列を生成
        while True:
            if (attacking_id := create_random_basetable_str(15)) in attacked_list:
                continue
            else:
                break

        if is_exist_seal_id(attacking_id):
            print(attacking_id)
            print(count)
            break
        else:
            attacked_list.append(attacking_id) 
            count += 1
        
    print("end")

attack_seal_id()
```
