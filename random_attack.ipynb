```python
import random

# ID
BASE_TABLE = ["F","S","L","G","J","X","R","K","C","I","Q","Z","V","A","M","W","U","N","T","B","E","H","P","O","Y","D"]

# 26()10
def id_to_decimal(src):
    # 
    reverse_src = ''.join(list(reversed(src)))
    
    dst = 0
    count = 0
    for letter in reverse_src:
        dst += BASE_TABLE.index(letter) * len(BASE_TABLE) ** count
        count += 1
    return dst

# ID
def is_exist_seal_id(id):
    # 10
    decimal_id = id_to_decimal(id)

    # 
    id_data = str(decimal_id)[:-2]
    checksum = str(decimal_id)[-2:]

    sum = 0
    for letter in str(decimal_id)[:-2]:
        sum += int(letter)
    
    print("==========")
    print(decimal_id)
    print(id_data)
    print(checksum)
    print(sum)
    
    if sum != checksum:
        return False
    
    # 1014

    # 5ID

    # OKID
    return True

# BASE_TABLE
def create_random_basetable_str(length):
    return "".join([BASE_TABLE[random.randrange(len(BASE_TABLE))] for length in range(length)])

# 
def attack_seal_id():
    count = 0
    attacked_list = []

    # TODO:
    while count < 10000:

        # 
        while True:
            if (attacking_id := create_random_basetable_str(15)) in attacked_list:
                continue
            else:
                break

        if is_exist_seal_id(attacking_id):
            print(attacking_id)
            print(count)
            break
        else:
            attacked_list.append(attacking_id) 
            count += 1
        
    print("end")

attack_seal_id()
```
